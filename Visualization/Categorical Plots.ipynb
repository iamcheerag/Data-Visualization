# Categorical Data Plots

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

They are basically used to compare b/w one numerical and one categorical column

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='size',y='tip',data = df)
```

```python
df.head()
```

```python
sns.countplot(x='smoker',data= df)
```

```python
sns.boxplot(x='size',y='total_bill',hue="smoker",data=df,palette='Set1')
```

```python
df.columns
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue='smoker',palette='plasma',split = True)
```

```python
sns.factorplot(x='day',y='total_bill',data=df,kind='swarm',hue='smoker')
```
